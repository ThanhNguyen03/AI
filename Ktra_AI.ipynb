Bài 1.1: SỬ DỤNG THUẬT TOÁN BACKTRACK VỚI KĨ THUẬT AC3.

```python
from simpleai.search import CspProblem, backtrack



# Định nghĩa biến và miền giá trị
#variables = provinces  # Mỗi tỉnh là một biến
#domains = {province colors for province in provinces}  # Miền giá trị của mỗi biến là các màu

# Định nghĩa ràng buộc
def constraint_func(provinces, values):
    # Ràng buộc: Hai tỉnh liền kề không thể có cùng màu
    return values[0] != values[1]
# Danh sách các tỉnh và các màu
provinces = (
    'An Giang', 'Đồng Tháp', 'Long An', 'Tiền Giang', 'Bến Tre', 'Vĩnh Long', 'Cần Thơ',
    'Trà Vinh', 'Kiên Giang', 'Hậu Giang', 'Sóc Trăng', 'Bạc Liêu', 'Cà Mau'
)
colors = dict((province,['Xanh', 'Đỏ', 'Vàng','Đen']) for province in provinces )

constraints = [
    (('An Giang', 'Đồng Tháp'), constraint_func),
    (('An Giang', 'Kiên Giang'), constraint_func),
    (('An Giang', 'Cần Thơ'), constraint_func),
    (('Đồng Tháp', 'Long An'), constraint_func),
    (('Đồng Tháp', 'Cần Thơ'), constraint_func),
    (('Đồng Tháp', 'Tiền Giang'), constraint_func),
    (('Long An','Tiền Giang'), constraint_func),
    (('Tiền Giang','Bến Tre'), constraint_func),
    (('Tiền Giang','Vĩnh Long'), constraint_func),

    (('Tiền Giang','Bến Tre'), constraint_func),
    (('Bến Tre','Vĩnh Long'), constraint_func),
    (('Bến Tre','Trà Vinh'), constraint_func),
    (('Vĩnh Long','Cần Thơ'), constraint_func),
    (('Vĩnh Long','Hậu Giang'), constraint_func),
    (('Vĩnh Long','Trà Vinh'), constraint_func),
    (('Vĩnh Long','Sóc Trăng'), constraint_func),
    (('Cần Thơ','Kiên Giang'), constraint_func),
    (('Cần Thơ','Hậu Giang'), constraint_func),

    (('Kiên Giang','Hậu Giang'), constraint_func),
    (('Kiên Giang','Bạc Liêu'), constraint_func),
    (('Kiên Giang','Cà Mau'), constraint_func),
    (('Hậu Giang','Sóc Trăng'), constraint_func),
    (('Hậu Giang','Bạc Liêu'), constraint_func),
    (('Sóc Trăng','Trà Vinh'), constraint_func),
    (('Sóc Trăng','Bạc Liêu'), constraint_func),
    (('Bạc Liêu','Cà Mau'), constraint_func)
    ]


# Tạo bài toán CSP
problem = CspProblem(provinces, colors, constraints)

# Giải bài toán bằng thuật toán backtracking
result = backtrack(problem)

# Kết quả
if result is not None:
    print("Kết quả:")
    for province, color in result.items():
        print(f"{province}: {color}")
else:
    print("Không tìm thấy giải pháp.")

```

```python
import string

# Đoạn văn bản đầu vào
text = """Tôi yêu truyện cổ nước tôi
Vừa nhân hậu lại tuyệt vời sâu xa
Thương người rồi mới thương ta
Yêu nhau dù mấy cách xa cũng tìm
Ở hiền thì lại gặp hiền
Người ngay thì gặp người tiên độ trì
Mang theo truyện cổ tôi đi
Nghe trong cuộc sống thầm thì tiếng xưa
Vàng cơn nắng, trắng cơn mưa
Con sông chảy có rặng dừa nghiêng soi
Đời cha ông với đời tôi
Như con sông với chân trời đã xa
Chỉ còn truyện cổ thiết tha
Cho tôi nhận mặt ông cha của mình
Rất công bằng, rất thông minh
Vừa độ lượng lại đa tình, đa mang.
Thị thơm thì giấu người thơm
Chăm làm thì được áo cơm cửa nhà
Đẽo cày theo ý người ta
Sẽ thành khúc gỗ chẳng ra việc gì
Tôi nghe truyện cổ thầm thì
Lời cha ông dạy cũng vì đời sau.
Đậm đà cái tích trầu cau
Miếng trầu đỏ thắm nặng sâu tình người.
Sẽ đi qua cuộc đời tôi
Bấy nhiêu thời nữa chuyển dời xa xôi.
Nhưng bao truyện cổ trên đời
Vẫn luôn mới mẻ rạng ngời lương tâm."""

# Bước 1: Chuyển đổi văn bản thành chữ thường
text = text.lower()

# Bước 2: Loại bỏ dấu câu và ký tự đặc biệt
text = ''.join([char for char in text if char not in string.punctuation])

# Bước 3: Tách văn bản thành các từ (tokens) dựa trên khoảng trắng
words = text.split()

# Bước 4: Loại bỏ các từ ngừng (stop words) nếu cần thiết
# stop_words = ["tôi", "lại", "mới", "dù", "cách", "trong", "có", "với", "của", "ở", "lại", "còn", "cho", "cùng", "thành", "và"]
# filtered_words = [word for word in words if word not in stop_words]

# Tạo danh sách từ điển (vocabulary) từ danh sách các từ đã tiền xử lý
vocabulary = list(set(words))

# Tạo vector BOW bằng cách đếm tần số xuất hiện của từng từ trong từ điển
bow_vector = []
for word in vocabulary:
    word_count = words.count(word)
    bow_vector.append(word_count)

# In vector BOW
print("Vector BOW:")
print(bow_vector)

# In từ điển (vocabulary) và vector BOW
print("Tần suất xuất hiện:")
for word, count in zip(vocabulary, bow_vector):
    print("Từ:", word, "\tTần số xuất hiện:", count)
```

Bài 2.2: SỬ DỤNG COUNTVECTORIZER CỦA SCIKIT-LEARN.

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from pyvi import ViTokenizer  # Sử dụng tokenizer tiếng Việt từ pyvi

# Văn bản tiếng Việt của bạn
vanban = """
Tôi yêu truyện cổ nước tôi
Vừa nhân hậu lại tuyệt vời sâu xa
Thương người rồi mới thương ta
Yêu nhau dù mấy cách xa cũng tìm
Ở hiền thì lại gặp hiền
Người ngay thì gặp người tiên độ trì
Mang theo truyện cổ tôi đi
Nghe trong cuộc sống thầm thì tiếng xưa
Vàng cơn nắng, trắng cơn mưa
Con sông chảy có rặng dừa nghiêng soi
Đời cha ông với đời tôi
Như con sông với chân trời đã xa
Chỉ còn truyện cổ thiết tha
Cho tôi nhận mặt ông cha của mình
Rất công bằng, rất thông minh
Vừa độ lượng lại đa tình, đa mang.
Thị thơm thì giấu người thơm
Chăm làm thì được áo cơm cửa nhà
Đẽo cày theo ý người ta
Sẽ thành khúc gỗ chẳng ra việc gì
Tôi nghe truyện cổ thầm thì
Lời cha ông dạy cũng vì đời sau.
Đậm đà cái tích trầu cau
Miếng trầu đỏ thắm nặng sâu tình người.
Sẽ đi qua cuộc đời tôi
Bấy nhiêu thời nữa chuyển dời xa xôi.
Nhưng bao truyện cổ trên đời
Vẫn luôn mới mẻ rạng ngời lương tâm
"""

# Sử dụng tokenizer tiếng Việt để chia văn bản thành các từ
tokenized_vanban = ViTokenizer.tokenize(vanban)

# Sử dụng CountVectorizer để biểu diễn tài liệu - từ
count_vectorizer = CountVectorizer()
document_term_matrix = count_vectorizer.fit_transform([tokenized_vanban])

# Lấy danh sách từ vựng
vocabulary = np.array(count_vectorizer.get_feature_names_out())

# Hiển thị kết quả thống kê tần số xuất hiện của từng từ
word_frequencies = document_term_matrix.toarray().flatten()
for word, freq in zip(vocabulary, word_frequencies):
    print(f'{word} : {freq}')
```
